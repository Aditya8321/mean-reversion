
# 01 - Data Collection
This notebook downloads historical stock data for our mean-reversion pairs and cleans it for analysis.


```python
import yfinance as yf
import pandas as pd
import os
```

```python
# Download data
tickers = ["KO", "PEP"]
data = yf.download(tickers, start="2010-01-01", end="2025-01-01")["Close"]

# Fill missing values
data = data.ffill().bfill()
```

```python
os.makedirs("../data/raw_prices", exist_ok=True)
os.makedirs("../data/processed", exist_ok=True)
```

```python

# Save raw and processed data
data.to_csv("../data/raw_prices/ko_pep.csv")
data.to_csv("../data/processed/ko_pep_clean.csv")

# Plot prices
import matplotlib.pyplot as plt
data.plot(figsize=(12,6), title="KO vs PEP Prices")
plt.show()
```
